```python
covid_df['positive_rate'] = covid_df.new_cases / covid_df.new_tests
```

```python
covid_df
```

```python
#inplace=True: This parameter is set to True to perform the operation in place, meaning that it will modify the original DataFrame without the need to create a new one.
covid_df.drop(columns = ['positive_rate'] , inplace = True)
```

```python
# uses quicksort sorting
# descending order
covid_df.sort_values('new_cases' , ascending = False).head(10)
```

```python
# ascending order
covid_df.sort_values('new_cases').head(10)
```

```python
covid_df.at[172, 'new_cases']= (covid_df.at[171, 'new_cases'] + covid_df.at[173, 'new_cases'])/ 2
```

```python
# coverting the date from datatype - object to datetime.
covid_df['date'] = pd.to_datetime(covid_df.date)
```

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```

```python
covid_df
```

```python
# query the rows for may
covid_df_may = covid_df[covid_df.month == 5]

# selecting the coloumns we want 
covid_df_may_metrices = covid_df_may[['new_cases', 'new_deaths', 'new_tests']]

# get the column-wise sum
covid_may_totals = covid_df_may_metrices.sum()
```

```python
# average for sundays
covid_df[covid_df.weekday == 6].new_cases.mean()
```

```python
covid_df_month = covid_df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()
```

```python
covid_df_month
```

```python
# to calculate the sum of cases upto a particular date.
covid_df['total_deaths'] = covid_df.new_deaths.cumsum() 
```

```python
merged_df = covid_df.merge(locations_df, on="location")
```

```python
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
```

```python
# writing data back to files
result_df = merged_df[[ 'date',
                       'new_cases',
                       'new_deaths',
                       'total_deaths', 
                       'new_tests', 
                       'deaths_per_million']]
```

```python
# to write the data from a dataframe into a file
# pandas turns index number also to a column in the file, so we use index =None to indicate that there is no need to include that.
result_df.to_csv('results.csv', index=None)
```
